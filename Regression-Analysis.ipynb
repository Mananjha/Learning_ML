```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate the data
x= 10* np.random.rand(50)
y= 2*x-5+np.random.rand(50)*4
```

```python
# Fit a linear regression
x=x.reshape(-1,1)
model= LinearRegression()
model.fit(x,y)
```

```python
#calculate the predicted values
y_pred=model.predict(x)
```

```python
# plot the scatter and regression line
plt.scatter(x,y, label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot with increased variability and regression lines")
```

```python
# plot the actual population line
x_line=np.linspace(0,10,100)
y_actual=2*x_line-5
plt.plot(x_line,y_actual, 'r',label="Population Line (m=2,b=-5)")

#plot the estimate regression line 
y_estimated= model.coef_[0]* x_line+model.intercept_
plt.plot(x_line,y_estimated,'g',label="Estimated Line(m={model.coef_[0]:.2f},b={model.intercept_:.2f})")

#add legend and show plot
plt.legend()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
url = "https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv"
data = pd.read_csv(url, index_col=0)

# Define independent variables (features)
X = data[['TV', 'Radio', 'Newspaper']]

# Define dependent variable (target)
y = data['Sales']

# Add a constant to the model (for intercept)
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
